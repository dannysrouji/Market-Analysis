```python
## classify the months into numerical months
def month_class(mon):
    if mon == 'jan' :
        return 1
    elif mon == 'feb':
        return 2
    elif mon == 'mar':
        return 3
    elif mon == 'apr':
        return 4
    elif mon == 'may':
        return 5
    elif mon == 'jun':
        return 6
    elif mon =='jul':
        return 7
    elif mon == 'aug':
        return 8
    elif mon =='sep':
        return 9
    elif mon == 'oct':
        return 10
    elif mon == 'nov':
        return 11
    else :
        return 12
```

```python
df['Month'] = df['month'].apply(month_class)
```
